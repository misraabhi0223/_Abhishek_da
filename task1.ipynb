LOAD THE DATASET "PENGUINS" FROM SEABORN LIBRARY AND THEN CLEAN IT

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['sex'].value_counts()
```

```python
df['bill_length_mm'].fillna(df['bill_length_mm'].mean(), inplace=True)
df['bill_depth_mm'].fillna(df['bill_depth_mm'].mean(), inplace=True)
df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean(), inplace=True)
df['body_mass_g'].fillna(df['body_mass_g'].mean(), inplace=True)
```

```python
df['sex'].fillna(df['sex'].value_counts().idxmax(), inplace=True)
```

```python
df.rename(columns={
    'bill_length_mm': 'length',
    'bill_depth_mm': 'depth', 
    'flipper_length_mm': 'flipper', 
    'body_mass_g': 'weight'},
    inplace=True)
df
```
