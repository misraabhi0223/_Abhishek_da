```python
import pandas as pd
import numpy as np
```

```python
# read data
df = pd.read_csv('mobiles.csv')
df.info()
df
```

```python
# processor function
def clean_price_data(price):
    if isinstance(price, str):
        price = price.replace(',','').replace('₹','').strip()
    if price.isnumeric():
        return float(price)
```

```python
def split_rw_data(rw):
    if isinstance(rw, str):
        rw = rw.split('&')
        return rw[0], rw[1]
    else:
        return np.nan, np.nan
```

```python
df['cleaned_price']=df['price'].apply(clean_price_data)
```

```python
df['ratings'], df['reviews'] = zip(*df['rw'].apply(split_rw_data))      # unpacking the tuple
```
